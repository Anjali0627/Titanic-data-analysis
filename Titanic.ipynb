```python

sex=pd.get_dummies(titanic_data['Sex'], drop_first=True)
embark=pd.get_dummies(titanic_data['Embarked'], drop_first=True)
```

```python
sex.head()
```

```python
embark.head()
```

```python
pcl=pd.get_dummies(titanic_data['Pclass'],drop_first=True)
```

```python
pcl.head()
```

```python
titanic_data.drop(['Sex', 'Pclass','Embarked', 'PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data=pd.concat([titanic_data, sex, embark, pcl], axis=1)
```

```python
titanic_data.head()
```

```python
X=titanic_data.drop('Survived', axis=1)
y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split from the correct modulet
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
# Handle non-numeric values in X_train
X_train = X_train.apply(pd.to_numeric, errors='coerce')
X_train = X_train.fillna(0)  # Replace missing values created by coercion with 0

# Now fit the model
logmodel.fit(X_train, y_train)
```

```python
# Handle non-numeric values in X_test as well
X_test = X_test.apply(pd.to_numeric, errors='coerce')
X_test = X_test.fillna(0)  # Replace missing values created by coercion with 0

# Now try predicting again
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predictions)
```
